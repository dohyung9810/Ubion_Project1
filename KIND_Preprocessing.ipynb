```python
import pandas as pd

def filter_spec_rows(df, column_name):
    # 특정 열에 "스팩"이라는 단어가 포함되지 않은 행만 선택
    filtered_df = df[~df[column_name].str.contains('스팩')]

    return filtered_df
# 'Column1' 열에 '스팩'이라는 단어가 들어간 행을 삭제
filtered_df = filter_spec_rows(df, '회사명')

# 결과 출력
filtered_df
```

```python
filtered_df['청구일_연도'] = filtered_df['청구일'].dt.year
```
