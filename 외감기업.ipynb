```python
def seperate_finance_guild_line(df):
  ifrs_use_com = df[ifrs].dropna()['거래소코드'].unique()
  ifrs_link_use_com = df[ifrs_link].dropna()['거래소코드'].unique()

  # 두 배열 간의 중복된 값 찾기
  common_cop = np.intersect1d(ifrs_use_com, ifrs_link_use_com)

  # print("두 배열 간의 중복된 값:", common_cop)
  print(len(common_cop))

  # 전체 기업 정보
  all_cop = df['거래소코드'].unique().tolist()

  mask = np.isin(ifrs_use_com, common_cop)
  ifrs_cop = ifrs_use_com[~mask].tolist()
  mask2 = np.isin(ifrs_link_use_com, ifrs_cop)
  ifrs_link_cop = ifrs_link_use_com[~mask2].tolist()

  mask3 = [name for name in all_cop if name not in ifrs_cop and name not in ifrs_link_cop]
  gaap_cop = mask3

  commom_fe = col_name[:36]
  ifrs.extend(commom_fe)
  ifrs_link.extend(commom_fe)
  gaap.extend(commom_fe)

  gaap_cop_df = df[df['거래소코드'].isin(gaap_cop)][gaap].reset_index(drop=True)
  gaap_cop_df['회계기준'] = 0

  ifrs_cop_df = df[df['거래소코드'].isin(ifrs_cop)][ifrs].reset_index(drop=True)
  ifrs_cop_df['회계기준'] = 1

  ifrs_link_cop_df = df[df['거래소코드'].isin(ifrs_link_cop)][ifrs_link].reset_index(drop=True)
  ifrs_link_cop_df['회계기준'] = 2

  rename_col = ['자산',
  '자본',
  '부채',
  '매출액',
  '법인세비용차감전순이익',
  '계속영업이익',
  '당기순이익',
  '투자활동으로 인한 현금흐름',
  '거래소코드']
  commom_fe.append('회계기준')
  rename_col.extend(commom_fe)

  ifrs_cop_df.columns = rename_col
  ifrs_link_cop_df.columns = rename_col
  gaap_cop_df.columns = rename_col

  df2 = pd.concat([ifrs_cop_df, ifrs_link_cop_df, gaap_cop_df], axis=0)
  # 중복된 열 확인
  duplicate_columns = df2.columns[df2.columns.duplicated()]

  df2 = df2.loc[:, ~df2.columns.duplicated()]

  df3 = df2.dropna()

  return df3
```
